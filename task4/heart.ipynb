## Data Preprocessing

```python
## lets rename the columns for better understanding
data = data.rename(columns={
    'cp': 'chest_pain_type',
    'trestbps': 'resting_blood_pressure',
    'chol': 'serum_cholesterol',
    'fbs': 'fasting_blood_sugar',
    'restecg': 'resting_ecg',
    'thalach': 'max_heart_rate',
    'exang': 'exercise_induced_angina',
    'oldpeak': 'st_depression',
    'slope': 'st_slope',
    'ca': 'major_vessels',
    'thal': 'thalassemia_type',
    'target': 'presence_of_heart_disease'
})
```

```python
data.columns
```

```python
## lets replace coded categorical values with their actual medical names

data['sex'] = data['sex'].map({0:'female',1:'male'})

data['chest_pain_type'] = data['chest_pain_type'].map({
    0: 'typical angina', 1: 'atypical angina',
    2: 'non-anginal pain', 3: 'asymptomatic'})

data['fasting_blood_sugar'] = data['fasting_blood_sugar'].map({
    0: 'less than 120 mg/dl', 1: 'greater than 120 mg/dl'})

data['resting_ecg'] = data['resting_ecg'].map({
    0: 'normal', 1: 'ST-T wave abnormality',
    2: 'left ventricular hypertrophy'})

data['exercise_induced_angina'] = data['exercise_induced_angina'].map({
    0: 'no', 1: 'yes'})

data['st_slope'] = data['st_slope'].map({
    0: 'upsloping', 1: 'flat', 2: 'downsloping'})

data['thalassemia_type'] = data['thalassemia_type'].map({
    1: 'normal', 2: 'fixed defect', 3: 'reversible defect'})

data['presence_of_heart_disease'] = data['presence_of_heart_disease'].map({
    0: 'no disease', 1: 'disease'})
```

## Preprocessing Pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


X = data.drop(columns=['presence_of_heart_disease'])
y = data['presence_of_heart_disease']

encode = LabelEncoder()
y_encode = encode.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encode,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object']).columns.tolist()

# Define Pipelines for preprocessing
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder())])

# Combine preprocessing steps
preprocessor = ColumnTransformer([
    ('num_features', num_pipeline, num_cols),
    ('cat_features', cat_pipeline, cat_cols)
])
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,precision_score, recall_score

lor = LogisticRegression()
lor.fit(X_train, y_train)

y_pred = lor.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print('='*50)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print('='*50)

print("Precision:", precision_score(y_test,y_pred))
print('Recall:',recall_score(y_test,y_pred))
print('='*50)
      
print("Classification Report:\n", classification_report(y_test, y_pred))
```

### Interpretation of the evaluation metrics

* **True Positives (TP) = 25**
  - Model correctly predicted 25 patients with heart disease.

* **False Negatives (FN) = 4**
  - 4 patients actually had heart disease, but the model predicted no disease → Risky in medical use cases.

* **Precision**:	80.6% → Of all predicted "heart disease" cases, ~81% were correct.

* **Recall**:	86.2% → Of all actual heart disease patients, the model caught ~86%.


### Sigmoid Function:

* Logistic Regression uses the sigmoid function to convert raw scores into probabilities.
By default, a threshold of 0.5 is used to classify outcomes, but this can be tuned.
Lowering the threshold increases recall (fewer false negatives), while raising it boosts precision (fewer false positives).

* We can tune this threshold based on the business need — for example, in heart disease detection, we may prefer higher recall to avoid missing high-risk patients.

```python
# Train the model
model2 = LogisticRegression()
model2.fit(X_train, y_train)

# Predict probabilities
y_proba = model2.predict_proba(X_test)[:, 1]

# Try a custom threshold
threshold = 0.4  #  change this to see effects
y_pred_custom = (y_proba >= threshold).astype(int)

# Evaluation

print("Accuracy:",accuracy_score(y_test,y_pred_custom))
print("Precision:",precision_score(y_test,y_pred_custom))
print("Recall:",recall_score(y_test,y_pred_custom))
print('='*50)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_custom))
print('='*50)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_custom))
```

### What Happened When the Threshold Lowered to 0.4?

* More samples were classified as class 1 (heart disease) because even lower probabilities were accepted.

* **Increased recall:**  The model is catching more positive cases (less likely to miss patients with heart disease).

* **Slight drop in precision:** It made a few more false positives (i.e., some healthy people classified as sick).

* Overall F1 score is balanced and good (0.84)
